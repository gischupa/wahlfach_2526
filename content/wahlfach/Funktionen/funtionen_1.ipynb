# Funktionen
Bisher hast du dein Programm immer so geschrieben, dass es Zeile für Zeile  
ausgeführt worden ist. Für sehr einfache Probleme ist das völlig in Ordnung.  
Steigt die Schwierigkeit des Problems, oder müssen Teile des Programms öfter  
ausgeführt werden, so soltlest du auf eine *Baustein-Technik* umsteigen.
Beim Programmieren nennt man diese Bausteine *Funktionen*.

> Die genaue Bezeichnung hängt von der Programmiersprache oder der  
> eingesetzten Technik ab. Manchmal wird auch von *Methoden* oder *Prozeduren*  
> gesprochen - im Kern sind das aber auch Funktionen!  

## Wie sieht eine Funktion aus?

Es gibt verschiedene Varianten, die du Schritt für Schritt lernen wirst.  
Die Varianten reichen von *dumme* Funktion bis *sehr schlaue* Funktion ...  

```python
def funktionsname():
    befehl_1
    befehl_2
    ...
```

(Ja - diese Funktion ist sinnlos! Aber wir bauen sie noch aus!)

#### Lösung    

````python
def mittelwert():
    print(  (5+6)/2  )

# Ausführung
mittelwert()
````

### Die Funktion etwas intelligenter machen
So ist die Funktion ist nicht wirklich hilfreich.   
Wie wäre es mit einer Funktion, die den Mittelwert von  
zum Beispiel 3 beliebigen Zahlen berechnen könnte?

Dafür brauchst du eine minimal andere Schreibweise:

```python
def mittelwert(a, b, c):
    print(   (a + b + c) / 3 )

# Ausführung mir strg + enter
mittelwert(7, 523, 14)
```

# Tic Tac Toe
Wenn wir schon Spiele prgrammieren, dann können wir auch ein einfaches  
Tic Tac Toe schreiben und dabei neue Techniken lernen!  
Wie könnte man ein Spielfeld auf den Bildschirm drucken?

#### Lösung

````python
print("   |   |   ")
print(" X | X | O ")
print("___|___|___")
print("   |   |   ")
print("   | X |   ")
print("___|___|___")
print("   |   |   ")
print(" O | X | X ")
print("   |   |   ")
````

### Verbessern der Ausgabe
In meiner Lösung sind dir bestimmt Zeilen aufgefallen, die sich wiederholen.  
Deine Aufgabe ist es nun, diese Ausgabe durch Funktionen etwas *schreibfreundlicher*  
zu gestalten. Ich hätte gerne die Möglichkeit, das Spielfeld so zu schreiben:

```python
def spielfeld():
    leerzeile()
    spielzeile("x","x","0")
    trennzeile()
    leerzeile()
    spielzeile(" ","x"," ")
    trennzeile()
    leerzeile()
    spielzeile("O","x","x")
    leerzeile()

# Ausführen
spielfeld()
```
