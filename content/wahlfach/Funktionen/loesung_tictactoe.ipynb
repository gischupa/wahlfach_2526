# Mögliche Lösung

Auch hier ist der Code auf verschiedene Felder verteilt, damit du nicht 
versehentlich die ganze Lösung siehst, bevor du das willst.

## Definition des Boards

````python
board = [
    ["-","-","-"],
    ["-","-","-"],
    ["-","-","-"],
]
````

## Anzeigen des Boards

````python
def zeichne_board(board):
    for row in range(3):
        print(board[row])
````

## Setzen eines Feldes

````python
def setze_feld(symbol, location, board):
    row = (location -1) // 3
    col = (location - 1)  % 3
    board[row][col] = symbol
````

## Noch Zellen frei?

````python
def noch_zellen_frei(board):
    if "-" in board[0] or "-" in board[0] or "-" in board[0]:
        return True
    return False
````

## Zeile voll?

````python
# Ist eine Zeile komplett mit x oder o belegt?
def zeile_voll(board, n):
    if board[n].count("x") == 3 or board[n].count("o") == 3:
        return True
    return False
    
````

## Spalte voll?

````python
def spalte_voll(board, n):
    spalte = []
    for i in range(3):
        spalte.append(board[i][n])
    if spalte.count("x") == 3 or spalte.count("o") == 3:
        return True
    return False
````

Vielleicht ist dir aufgefallen, dass wir nun schon 2x den gleichen Code geschrieben haben?  
Jedes mal haben wir geprüft, ob in einer Liste drei Kreuze oder 3 Kringel sind.  
Das lösen wir besser, durch eine eigene Funktion:

```python
def drei_in_liste(liste):
    if liste.count("x") == 3 or liste.count("o") == 3:
        return True
    return False


def zeile_voll(board, n):
    liste = board[n]
    return drei_in_liste(liste)


def spalte_voll(board, n):
    liste = []
    for i in range(3):
        liste.append(board[i][n])
    return drei_in_liste(liste)
```

Nach diesem System kannst du auch leicht `diagonale_links_voll`  
und `diagonale_rechts_voll` schreiben.
                     

## Diagonalen voll?

````python
def diagonale_links_voll(board):
    liste = []
    liste.append(board[0][0])
    liste.append(board[1][1])
    liste.append(board[2][2])
    return drei_in_liste(liste)

def diagonale_rechts_voll(board):
    liste = []
    liste.append(board[0][2])
    liste.append(board[1][1])
    liste.append(board[2][1])
    return drei_in_liste(liste)
````

## Gewonnen?

Hier kommt nun alles zusammen. Zuerst alle Zeilen, dann die Diagonalen prüfen.  

**WICHTIG**  
Damit der Test auf einen Sieg möglichst übersichtlich bleibt, solltest du dir folgende Beziehungen merken:
```
True and True = True
True and False = False
False and False = False
```

Bei oder (`or`) sieht das anders aus:
```
True or True = True
True or False = True
False or False = False
```

Nun musst du alle möglichen Varianten durchtesten. Hilfreich können hier zusätzliche Funktionen sein:

* `zeilen_voll`  (beachte den Plural)
* `spalten_voll`
* `diagonalen_voll`

````python
def zeilen_voll(board):
    voll = False   # IRGENDEINE zeile genügt
    voll = voll or zeile_voll(board, 0)
    voll = voll or zeile_voll(board, 1)
    voll = voll or zeile_voll(board, 2)
    return voll
````

````python
def spalten_voll(board):
    voll = False   # IRGENDEINE Spalte genügt
    voll = voll or spalte_voll(board, 0)
    voll = voll or spalte_voll(board, 1)
    voll = voll or spalte_voll(board, 2)
    return voll
````

````python
def diagonalen_voll(board):
    voll = False
    voll = voll or diagonale_links_voll(board)
    voll = voll or diagonale_rechts_voll(board)
    return voll
````

````python
def gewinn(board):
    gewinn = False
    gewinn = gewinn or zeilen_voll(board)
    gewinn = gewinn or spalten_voll(board)
    gewinn = gewinn or diagonalen_voll(board)
    return gewinn
````

# Spielschleife
Der Computer muss ständig zwischen den beiden „Spielern“ (also den Symbolen)
hin und her schalten. Nach jedem Zug muss auf Gewinn oder unentschieden geprüft werden.

````python
def switch_symbol(symbol):
    if symbol == "x":
        symbol = "o"
        return symbol
    symbol = "x"
    return symbol

def game_loop(board):
    game_is_running = True
    symbol = "x"
    while game_is_running:
        feld_nr = int(input("Bitte freies Feld wählen (1-9): "))
        setze_feld(symbol, feld_nr, board)
        if gewinn(board) is True:
            print("Gewonnen hat " + symbol)
            break
        if noch_zellen_frei(board) is False:
            print("Unentschieden")
        zeichne_board(board)    
        symbol = switch_symbol(symbol)
    
            
            
game_loop(board)        
                      
````
