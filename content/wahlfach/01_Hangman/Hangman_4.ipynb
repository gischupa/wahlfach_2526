# Die Eingabe des Benutzers
Hier kommt wieder der aktuelle Stand, damit du weiter machen kannst.

````python
from random import choice

# Wort wählen
zu_raten_pool = ["apfel", "pfirsich", "banane", "kirsche"]
zu_raten = choice(zu_raten_pool)

geraten = ["b", "a"]

aktuell = ""
for buchstabe in zu_raten:
    if buchstabe in geraten:
        aktuell += buchstabe + " "
    else:
        aktuell += "_ "
print(aktuell)
````

## Endlos-Schleife
Der Benutzer soll nun ständig Buchstaben eingeben. Über einen Abbruch beim komplett   
richtigen Wort denken wir dann später nach. Hier soll die Eingabe enden, wenn der Benutzer  
eine Ziffer zippt.  

Wie geht eine *Endlos-Schleife*?  
Auf alle Fälle nicht mit `for i in range(...)`, denn da kennen wir die  
Anzahl der Versuche. Also muss es mit `while` gemacht werden.

**Warnung**  
Probier das in dieser Form NICHT aus - dein Browser könnte abstürzen,  
weil du die Schleife nicht beenden kannst!  

```python
while True:
   mach etwas
```

#### Was soll der Benutzer machen? 
Natürlich etwas eingeben!

```python
while True:
   buchstabe = input("Neuer Buchstabe bitte. (Ziffer = Abbruch) ")
   geraten.append(buchstabe)
   ...
   if buchstabe in "0123456789":
       break
```

### Lösung 2 (Alle Buchstaben da)

````python
zu_raten = "banane"
geraten = ["b","a","n"] # das "e" fehlt also noch.

# Wort zerlegen in Buchstaben - Doppelte nehmen wir mit
buchstaben_pool = []
for buchstabe in zu_raten:
    buchstaben_pool.append(buchstabe)

# Übliche Rate-Darstellung
aktuell = ""
for buchstabe in zu_raten:
    if buchstabe in geraten:
        aktuell += buchstabe + " "
    else:
        aktuell += "_ "

# Check, ob alle aus dem "pool" in "geraten" sind
alle_da = True                     # Vorgabe, die wir evtl. ändern
for buchstabe in buchstaben_pool:  # Jeden Buchstaben testen
    if not buchstabe in geraten:   # Fehlt er?
        alle_da = False            # Vorgabe ändern = vorbei

# break Entscheidung treffen
if alle_da:
    print("break")
else:
    print("kein break")       
````
