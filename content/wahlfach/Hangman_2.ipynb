## Pakete
Python ist *klein* und kann durch Zusatzpakete erweitert werden.  
Diese Pakete musst du am Anfang des Programms *importieren*.  
Hier gibt es einige Varianten, die du an verschiedenen Stellen treffen wirst.  
Wir wollen aber aus dem Paket `random` lediglich den Befehl `choice` verwenden.  
Schreib also `from random import choice`

Nun geben wir dem Computer 4 Wörter vor. Die Wörter könnte man zum Beispiel  
auch aus einer Datei lesen - einem Wörterbuch zum Beispiel.  
Wir verwenden eine Liste:

```python
zu_raten_pool = ["apfel", "pfirsich", "banane", "kirsche"]
```

Aus diesem Pool soll nun ein zufälliges Wort gewählt werden (choice):

```python
zu_raten = choice(zu_raten_pool)
```
