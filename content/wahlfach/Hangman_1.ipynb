## Durch eine Zeichenkette laufen
Eine **wichtige Technik** beim Programmieren ist die **Wiederholung von Befehlen**.  
Es gibt dafür zwei Befehle, wir sehen uns aber nur die `for`-Schleife an:  

```python
for i in range(5):
    print("Zahl ", i)
```

Hier wird der eingerückte Befehl 5x ausgeführt (das sagt `range(5)`).  
Bei jedem Durchlauf besitzt `i` einen höheren Wert. Allerdings musst du  
wissen, dass die Nummerierung mit 0 beginnt und damit sind 5 Werte   
die Zahlen 0,1,2,3,4.

### Das hat das mit einer Zeichenkette zu tun?
Auf die einzelnen Buchstaben kannst du so zugreifen:

```python
wort = "Hallo"
print(wort[0])  # liefert "H"
print(wort[1])  # liefert "a"
print(wort[5])  # liefert einen Fehler!!!!
```

Das wollen wir ausnützen und durchlaufen "Hallo" mit einer Schleife.  
Probiere zunächst, ob das selbst schaffst, bevor du in die Lösung schaust"
