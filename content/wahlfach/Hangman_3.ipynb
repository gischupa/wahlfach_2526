# Rate-Darstellung verfeinern
Damit du auf dem Stand der letzten Stunde bist, habe ich in die nächste Codezeile  
den aktuellen Stand geschrieben:

````python
from random import choice

# Wort wählen
zu_raten_pool = ["apfel", "pfirsich", "banane", "kirsche"]
zu_raten = choice(zu_raten_pool)

# Rate-Darstellung erzeugen
aktuell = ""
for i in range(len(zu_raten)):
    aktuell += "_ "
print(aktuell)
````

## Vorarbeiten
Bevor wir zur eigentlichen Eingabe des Benutzers kommen, musst du dir überlegen,  
wie du den aktuellen Spielstand anzeigst. Angenommen das Wort war *Banane* und der  
Spieler hat die Buchstaben "b" und "a" geraten, dann muss auf dem Bildschirm stehen  

> b a _ a _ _

Du musst also richtige Buchstaben erkennen und im Wort `aktuell` passend ersetzen.  
Wenn du nicht weißt, wie du anfangen sollst - kein Problem. Wir machen ein paar  
Übungen, damit du dir leichter tust.  

Überlege dir zunächst, wie du die Eingaben des Benutzers „verwaltest“. Er gibt ja  
bei jedem Versuch einen neuen Buchstaben ein und den muss der Computer sich merken.  
Hast du eine Idee, wie das gehen könnte? Nimm an, der Benutzer hätte *b* und *a* geraten.

#### Lösung

````python
geraten = ["b", "a"] 
````

## Buchstabe schon geraten?
Nun musst du für jeden Buchstaben des geheimen Wortes prüfen,  
ob der aktuell Buchstabe in `geraten` vorkommt. Auch hier  
hat Python einen Befehl: `in`  
```python
if "a" in ["x","y","z"]:
    print("Buchstabe kommt vor")
else:
    print("Buchstabe kommt nicht vor")
```

Deine Aufgabe ist es nun 
> für jeden Buchstaben in `zu_raten`

nachzusehen, ob er in `geraten` vorkommt. Wenn ja, dann steht in `akutell` der Buchstabe,  
ansonsten, schreibst du wieder `_`. Schaffst du das? Probiere es wieder selbst, bevor zu in die  
Lösung schaust!

**Tipp**  
Du musst im Code den Teil mit *„Rate-Darstellung erzeugen“* umschreiben.   


#### Lösung

````python
geraten = ["b", "a"]  # kommt eigentlich vom Benutzer!

aktuell = ""
for buchstabe in zu_raten:
    if buchstabe in geraten:
        aktuell += buchstabe + " "
    else:
        aktuell += "_ "
print(aktuell)
````
